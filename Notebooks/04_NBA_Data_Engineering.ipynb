```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
df = pd.read_csv("all_seasons_clean.csv")
df.head()
```

```python
df.shape
df.info()
df.head()
```

```python
df = df.drop(columns=["player_name"])
```

```python
df["height_m"] = df["player_height"] / 100
df["BMI"] = df["player_weight"] / (df["height_m"] ** 2)
```

```python
df["season_start"] = df["season"].str[:4].astype(int)
draft_year_numeric = pd.to_numeric(df["draft_year"], errors="coerce")
df["experience"] = df["season_start"] - draft_year_numeric
```

```python
df["is_drafted"] = np.where(df["draft_year"] == "Undrafted",0,1)
```

```python
df["height_category"] = pd.cut(
    df["player_height"],
    bins=[170,190,205,240],
    labels=["Guard","Wing","Big"]
)
```

```python
df["weight_category"] = pd.cut(
    df["player_weight"],
    bins=[60,85,105,150],
    labels=["Light","Medium","Heavy"]
)
```

```python
df = pd.get_dummies(
    df,
    columns=[
        "team_abbreviation",
        "country",
        "height_category",
        "weight_category"
    ],
    drop_first=True
)
```

```python
df["draft_round"] = pd.to_numeric(
    df["draft_round"],
    errors="coerce"
)

df["draft_number"] = pd.to_numeric(
    df["draft_number"],
    errors="coerce"
)
```

```python
df.isnull().sum()
```

```python
df.to_csv("all_seasons_feature_engineered.csv", index=False)
print("Feature engineered dataset saved successfully!")
```
