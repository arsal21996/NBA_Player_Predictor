```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_csv("all_seasons.csv")

df.drop(columns="Unnamed: 0", inplace=True)
```

```python
clean_df = df.copy()
```

```python
clean_df["college"].isna().sum()
```

```python
clean_df["college"] = clean_df["college"].fillna("Unknown")
```

```python
duplicates = clean_df[
    clean_df.duplicated(
        subset=["player_name", "season"],
        keep=False
    )
]

duplicates
```

```python
draft_cols = [
    "draft_year",
    "draft_round",
    "draft_number"
]

for col in draft_cols:
    clean_df[col] = clean_df[col].replace(
        "Undrafted",
        np.nan
    )
```

```python
clean_df[draft_cols] = clean_df[draft_cols].apply(
    pd.to_numeric
)
```

```python
clean_df["gp"].describe()

clean_df["ts_pct"].describe()

clean_df["net_rating"].describe()
```

```python
clean_df["team_abbreviation"] = (
    clean_df["team_abbreviation"]
    .str.upper()
)
```

```python
clean_df["player_name"] = (
    clean_df["player_name"]
    .str.strip()
)
```

```python
clean_df.info()

clean_df.isna().sum()

clean_df.duplicated().sum()
```

```python
clean_df.to_csv(
    "all_seasons_clean.csv",
    index=False
)
```
