```python
# ======================================
# Correlation Analysis
# ======================================

# Compute correlation matrix
corr = df.select_dtypes(include=np.number).corr()

# Display correlation matrix
corr
## Heatmap
plt.figure(figsize=(12, 8))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Numerical Features")
plt.show()
## Strong Positive Correlations
# Strong positive correlations (> 0.6)
positive_corr = (
    corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
        .stack()
        .sort_values(ascending=False)
)

positive_corr[positive_corr > 0.60]
## Strong Negative Correlations
# Strong negative correlations (< -0.30)
negative_corr = (
    corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
        .stack()
        .sort_values()
)

negative_corr[negative_corr < -0.30]
## Short Interpretation (Markdown Cell)
## Observations:
## - Points (PTS) are expected to have a strong positive correlation with Usage Percentage (USG%).
## - Rebounds (REB) generally show a positive correlation with player height and weight.
## - Assists (AST) often correlate positively with Usage Percentage and AST%.
## - TS% may have a moderate positive relationship with Points.
## - Very strong negative correlations are generally uncommon in this dataset.
```

```python
## 7. Group-Based Analysis
## 7.1 Average Player Statistics by Season
season_stats = (
    df.groupby("season")[["pts", "reb", "ast", "age"]]
      .mean()
      .round(2)
)

season_stats
## 7.2 Average Points by Season
plt.figure(figsize=(12,5))

sns.lineplot(
    data=season_stats,
    x=season_stats.index,
    y="pts",
    marker="o"
)

plt.title("Average Points per Player by Season")
plt.xlabel("Season")
plt.ylabel("Average Points")
plt.xticks(rotation=45)

plt.show()
## 7.3 Average Rebounds by Season
plt.figure(figsize=(12,5))

sns.lineplot(
    data=season_stats,
    x=season_stats.index,
    y="reb",
    marker="o"
)

plt.title("Average Rebounds per Player by Season")
plt.xlabel("Season")
plt.ylabel("Average Rebounds")
plt.xticks(rotation=45)

plt.show()
## 7.4 Average Assists by Season
plt.figure(figsize=(12,5))

sns.lineplot(
    data=season_stats,
    x=season_stats.index,
    y="ast",
    marker="o"
)

plt.title("Average Assists per Player by Season")
plt.xlabel("Season")
plt.ylabel("Average Assists")
plt.xticks(rotation=45)

plt.show()
## 7.5 Average Age by Season
plt.figure(figsize=(12,5))

sns.lineplot(
    data=season_stats,
    x=season_stats.index,
    y="age",
    marker="o"
)

plt.title("Average Player Age by Season")
plt.xlabel("Season")
plt.ylabel("Average Age")
plt.xticks(rotation=45)

plt.show()
## 7.6 Average Points by Country (Top 10)

## Filter countries with at least 10 player-seasons to avoid small-sample outliers.

country_pts = (
    df.groupby("country")
      .agg(
          Avg_Points=("pts", "mean"),
          Players=("player_name", "count")
      )
)

country_pts = (
    country_pts[country_pts["Players"] >= 10]
    .sort_values("Avg_Points", ascending=False)
    .head(10)
)

country_pts
plt.figure(figsize=(10,5))

sns.barplot(
    data=country_pts,
    x="Avg_Points",
    y=country_pts.index
)

plt.title("Top 10 Countries by Average Points")
plt.xlabel("Average Points")
plt.ylabel("Country")

plt.show()
## 7.7 Average Points by Team
team_pts = (
    df.groupby("team_abbreviation")["pts"]
      .mean()
      .sort_values(ascending=False)
)

team_pts
plt.figure(figsize=(12,6))

sns.barplot(
    x=team_pts.values,
    y=team_pts.index
)

plt.title("Average Points by Team")
plt.xlabel("Average Points")
plt.ylabel("Team")

plt.show()
## 7.8 Key Grouped Tables
print("Season-wise Average Statistics")
display(season_stats)

print("\nTop Countries by Average Points")
display(country_pts)

print("\nTeams by Average Points")
display(team_pts.to_frame("Average Points"))
```

## Key Insights

### 1. Highest Average Scoring Season
- The **2022–23** season recorded the highest average points per player-season (**9.12 PPG**) in the dataset.
- Recent NBA seasons (2018–19 onward) generally show higher average scoring, reflecting the league's increasingly offense-oriented style of play.

### 2. Highest-Scoring Countries
- Among countries with at least 10 player-seasons, the **U.S. Virgin Islands** produced the highest average scoring players (**18.89 PPG**).
- Other high-scoring countries include **Bahamas**, **Cameroon**, **Serbia and Montenegro**, and the **United Kingdom**.
- The **United States** contributes the majority of players in the dataset, while several smaller countries have high averages due to fewer elite players.

### 3. Height vs. Rebounds
- The correlation between **player height** and **rebounds** is **0.424**, indicating a **moderate positive relationship**.
- Taller players generally collect more rebounds, although height alone does not fully determine rebounding ability.

### 4. Usage Percentage vs. Points
- The correlation between **Usage Percentage (USG%)** and **Points** is **0.641**, representing a **strong positive relationship**.
- Players with a larger offensive role tend to score significantly more points.

### 5. Highest-Scoring Teams
- Based on average player-season points, the top teams are:
  1. NOP
  2. GSW
  3. LAL
  4. PHX
  5. BKN
- These teams have had player-seasons with relatively high offensive production throughout the dataset.

### Overall Observation
- Offensive statistics have generally increased in recent NBA seasons.
- Usage Percentage is one of the strongest indicators of scoring performance.
- Physical attributes, particularly height, contribute positively to rebounding but have a weaker relationship than usage has with scoring.
