## Dataset Description

The dataset contains NBA player season statistics from the **1996–97** season through the **2022–23** season.

Each row represents one player's performance during a single NBA season and contains information about:

- Player information (name, age, height, weight)
- Team information
- Draft details
- College and country
- Performance statistics (points, rebounds, assists, games played)
- Advanced basketball metrics such as Net Rating, True Shooting Percentage (TS%), Usage Percentage (USG%), Offensive Rebound Percentage (OREB%), Defensive Rebound Percentage (DREB%), and Assist Percentage (AST%).

The dataset consists of **12,844 records** and **21 features**.

## Dataset Dimensions

- Total Rows: **12,844**
- Total Columns: **21**
- Unique Players: **2,551**
- Seasons Covered: **27**
- Teams Represented: **36**
- Countries Represented: **82**
- Colleges Represented: **356**

## Feature Types

The dataset contains two main types of variables.

### Numerical Features
These include continuous and discrete player statistics such as:

- Age
- Height
- Weight
- Games Played (GP)
- Points (PTS)
- Rebounds (REB)
- Assists (AST)
- Net Rating
- Offensive Rebound %
- Defensive Rebound %
- Usage %
- True Shooting %
- Assist %

### Categorical Features
These describe player identity and background:

- Player Name
- Team Abbreviation
- College
- Country
- Draft Year
- Draft Round
- Draft Number
- Season


## Missing Values Analysis

Only one feature contains missing values:

| Feature | Missing Values | Percentage |
|---------|---------------:|-----------:|
| College | 1,854 | 14.43% |

The missing college values are expected because many NBA players entered the league directly from high school or international leagues, making these missing values acceptable rather than data errors.


## Duplicate Records

No completely duplicated rows were found.

However, eight records share the same player name and season. These are not actual duplicates because they belong to different players with identical names (for example, two different players named Marcus Williams).

Therefore, no duplicate records were removed.

## Data Quality Checks

Several validation checks were performed on the dataset.

The following were verified:

- No negative values for points, rebounds, or assists.
- No players younger than 18 years.
- No unrealistic player heights or weights.
- No completely duplicated records.

Some extreme values were identified:

- Players with more than 82 games played.
- TS% values greater than 1.
- Extremely high and low Net Rating values.

These values mostly belong to players with very few games played, causing advanced statistics to become unstable. They were retained because they represent valid observations rather than data entry errors.


## Observations

Several interesting observations were identified during the exploratory analysis:

- Vince Carter appears in the largest number of seasons (22).
- Jeff Green and Mike James played for the highest number of NBA teams (11 each).
- Approximately 14% of players have missing college information.
- Undrafted players form the largest draft category.
- Most player statistics fall within realistic NBA ranges.
- Advanced statistics such as Net Rating and True Shooting Percentage contain extreme values caused by players with very limited playing time.

## Conclusion

The dataset is generally clean and suitable for analysis.

Apart from missing values in the College column and a few expected statistical outliers, no major data quality issues were found.

The dataset is ready for exploratory data analysis, visualization, feature engineering, and predictive modeling.
